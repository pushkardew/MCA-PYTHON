```python
# Display Average Salary by Department from CSV

import csv

file_name = "file.csv"
dept_salary = {}

try:
    with open(file_name, "r") as file:
        reader = csv.reader(file)
        for name, dept, sal in reader:
            sal = float(sal)
            dept_salary.setdefault(dept, []).append(sal)
    print("Average Salary by Department:")
    for dept, sals in dept_salary.items():
        print(f"{dept}: {sum(sals)/len(sals):.2f}")
except FileNotFoundError:
    print(f"Error: The file '{file_name}' was not found. Please make sure it's in the same directory.")
except ValueError:
    print("Error: Salary data might not be a valid number in the CSV file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
