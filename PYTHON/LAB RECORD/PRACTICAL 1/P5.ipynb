```python
# Vowel Reversal

user_input = input("Enter sentence: ")
print("Input:", user_input)
```

```python
def reverse_vowels(s):
    vowels = "aeiouAEIOU"
    s_list = list(s)  
    vowel_indices = [i for i, char in enumerate(s_list) if char in vowels]
    actual_vowels = [s_list[i] for i in vowel_indices]
    for i, original_index in enumerate(vowel_indices):
        s_list[original_index] = actual_vowels[len(actual_vowels) - 1 - i] 
    return "".join(s_list) 
result = reverse_vowels(user_input)
print("Output:", result)
```
