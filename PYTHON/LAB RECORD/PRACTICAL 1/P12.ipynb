```python
students = ["Leeladhar", "Aman", "Lokesh", "Mukesh", "Kunal", "Rahul"]

print("Student List:", students)
```

```python
import random

students = ["Leeladhar", "Aman", "Lokesh", "Mukesh", "Kunal", "Rahul"]

# Select CR randomly
cr = random.choice(students)
print("Selected CR:", cr)

# Shuffle the list
random.shuffle(students)
print("Shuffled Student List:", students)
```
