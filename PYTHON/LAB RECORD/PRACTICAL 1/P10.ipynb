```python
num = int(input("Enter a number: "))
power = int(input("Enter power value: "))

print("Number entered by user:", num)
print("Power entered by user:", power)
```

```python
import math

try:
    # a) Square root
    sqrt_value = math.sqrt(num)

    # b) Factorial
    fact_value = math.factorial(num)

    # c) Power
    power_value = math.pow(num, power)

    print("Square root of", num, "is:", sqrt_value)
    print("Factorial of", num, "is:", fact_value)
    print(num, "raised to power", power, "is:", power_value)

except Exception as e:
    print("Error:", e)
```
