```python
def print_pattern_a(rows):
    for i in range(1, rows + 1):
        for j in range(1, i + 1):
            print(j, end="")
        print()
print_pattern_a(5)
```

```python
def print_pattern_b(rows):
    for i in range(1, rows + 1):
        print("*" * i)

print_pattern_b(5)

```

```python
def print_pattern_c(rows):
    # Top half
    for i in range(rows):
        print(" " * (rows - i - 1) + "*" * (2 * i + 1))
    # Bottom half
    for i in range(rows - 2, -1, -1):
        print(" " * (rows - i - 1) + "*" * (2 * i + 1))

print_pattern_c(4)
```

```python
def print_pattern_d(rows,cols):
    for i in range(rows):
        for j in range(cols):
            # Check if the current position is on the border
            if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:
                print("*", end="")
            else:
                print(" ", end="")
        # Move to the next line after eachÂ row
        print()

print_pattern_d(6,6)

```

```python
def print_pattern_e(n):
    
    for i in range(n, 0, -1):
        print(' ' * (n - i), end='')
        print('* ' * i)

    for i in range(2, n + 1):
        print(' ' * (n - i), end='')
        print('* ' * i)
        
print_pattern_e(5)
```
