```python

print("Enter list item one by one and enter 'done' after completed")
list=[]
item=None
while(item!="done"):
    item=input()
    if(item=="done"):
        break
    list.append(item)
print(f"Entered list is:\n{list}")
```

```python
n=input("Enter the item to be searched:")
print(f"Entered item:{n}")
```

```python
if n in list:
    print(f"Item found at position {list.index(n) + 1}")
else:
    print("Item not found.")
```
